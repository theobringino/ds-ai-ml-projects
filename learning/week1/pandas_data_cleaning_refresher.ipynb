```python
#Changes and drops nulls from original dataframe
df = pd.read_csv('data.csv')
df.dropna(inplace = True)
df
```

```python
# replace all nulls with value
df = pd.read_csv('data.csv')
df.fillna(23, inplace = True)
df
```

```python
#replace nulls on specific columns only
df = pd.read_csv('data.csv')
df.fillna({"Calories" : 130}, inplace = True)
df
```

Imputation

```python
df = pd.read_csv('data.csv')
imp = df['Calories'].mean()
df.fillna({'Calories' : imp}, inplace = True)
df
```

```python
df = pd.read_csv('data.csv')
imp = df['Calories'].median()
df.fillna({'Calories' : imp}, inplace = True)
df
```

```python
df = pd.read_csv('data.csv')
imp = df['Calories'].mode()
df.fillna({'Calories' : imp}, inplace = True)
df
```
